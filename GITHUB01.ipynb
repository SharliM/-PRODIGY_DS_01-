```python
import pandas as pd

# Define the file path
file_path = 'kaggle.csv'

# Load the dataset into a DataFrame
df = pd.read_csv(file_path)

# Separate survival data by gender
male_survival = df[df['Sex'] == '1'].index
female_survival = df[df['Sex'] == '2'].index

# Perform one-way ANOVA
from scipy.stats import f_oneway

f_statistic, p_value = f_oneway(male_survival, female_survival)

# Print results
print("F-statistic:", f_statistic)
print("p-value:", p_value)

# Interpret the results
alpha = 0.05  # significance level
if p_value < alpha:
    print("Reject the null hypothesis. There is significant evidence that the survival rates are different between genders.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to conclude that the survival rates are different between genders.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset into a DataFrame
df = pd.read_csv('kaggle.csv')

# Count the number of occurrences for each category in the 'Sex' column
sex_counts = df['Sex'].value_counts()

# Plot a bar graph for gender distribution
plt.figure(figsize=(8, 6))
sex_counts.plot(kind='bar', color='skyblue')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)  # Rotate x-axis labels if needed
plt.show()
```

```python
import pandas as pd

# Load the dataset into a DataFrame
df = pd.read_csv('kaggle.csv')

# Convert non-numeric columns to numeric
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Drop rows with NaN values (if any)
df.dropna(inplace=True)

# Compute correlation matrix
correlation_matrix = df.corr()

# Print correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)

# Optional: Visualize correlation matrix
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```
